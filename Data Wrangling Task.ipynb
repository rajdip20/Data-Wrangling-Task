# b. Tidiness Issues

- Phone and email together in one column i.e. contact
- 3 variables stored in auralin and novodra column
- Merge the advreact table to treatment table

<b>Define</b>

Phone and email together in one column i.e. contact. so split it.

<b>Code</b>

```python
patients['email'] = patients['contact'].str.findall('[a-zA-Z_.]+@[a-zA-Z0-9-]+\.[a-zA-Z.]+')
```

```python
patients['phone']=patients['contact'].str.findall(r'[\+\(]?[1-9][0-9 .\-\(\)]{8,}[0-9]')
```

```python
patients.drop(['contact'],axis=1,inplace=True)
```

<b>Test</b>

```python
patients
```

<b>Define</b>

3 variables stored in auralin and novodra column. so splite it.

<b>Code</b>

```python
treatment = pd.melt(treatment,id_vars=['given_name','surname','hba1c_start','hba1c_end','hba1c_change'],var_name='treatment',value_name='dose')
```

```python
treatment = treatment[treatment['dose']!='-']
```

```python
treatment['low'],treatment['high'] = treatment['dose'].str.split('-').str
```

```python
treatment.drop(columns=['dose'],inplace=True)
```

<b>Test</b>

```python
treatment.head()
```

<b>Define</b>

Merge the advreact table to treatment table

<b>Code</b>

```python
treatment = pd.merge(treatment, advreact, how="left", on=['given_name','surname'])
```

```python
treatment.isnull().sum()
```

```python
treatment['adverse_reaction']=treatment['adverse_reaction'].fillna('No')
```

```python
def reaction(value):
  if value=='No':
    return 'No'
  else:
    return 'Yes'
```

```python
treatment['adverse_reaction']=treatment['adverse_reaction'].apply(reaction)
```

<b>Test</b>

```python
treatment.sample(10)
```

# c. Validity Issues
- Zip code column is stored in decimal format
- Certain Zip codes of 4 digits
- Novodra and Auralin columns has u attached to numbers
- Incorrect datatype for auralin and novodra columns

<b>Define</b>

Zip code column is stored in decimal format. it should be intiger.

<b>Code</b>

```python
patients['zip_code'] = patients['zip_code'].astype('int64')
```

<b>Test</b>

```python
patients.info()
```

Certain Zip codes of 4 digits. it should be five digit. (assuming the last digit is 0)

<b>Code</b>

```python
patients['zip_code']=patients['zip_code'].astype('str').str.pad(5,'right','0')
```

<b>Test</b>

```python
patients
```

Novodra and Auralin columns has u attached to numbers. (Now it becomes low and high)

```python
treatment['high']=treatment['high'].str.split('u').str[0]
```

<b>Test</b>

```python
treatment.head()
```

<b>Code</b>

```python
def capitalize(value):
    return value.title()
```

```python
patients['given_name']=patients['given_name'].apply(capitalize)
treatment['given_name']=treatment['given_name'].apply(capitalize)
advreact['given_name']=advreact['given_name'].apply(capitalize)
patients['surname']=patients['surname'].apply(capitalize)
treatment['surname']=treatment['surname'].apply(capitalize)
advreact['surname']=advreact['surname'].apply(capitalize)
```
